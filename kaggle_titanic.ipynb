```
import numpy as np
from keras.layers import Dense, Input, Flatten
from keras.models import Model
import pandas as pd
```

```
raw_input = pd.read_csv('/content/drive/My Drive/kaggle/input_data/train.csv')
raw_input_test = pd.read_csv('/content/drive/My Drive/kaggle/input_data/test.csv')
raw_input.head()
```

```
def CleanData(raw_input):
  raw_input['Sex'].replace({'male': 1, 'female': 0}, inplace=True)
  raw_input.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)
  raw_input=raw_input.astype({'Embarked': 'string'})
  raw_input['Embarked'].replace({'S': 1, 'C': 2, 'Q': 3}, inplace=True)
  raw_input=raw_input.astype({'Embarked': 'string'})
  raw_input['Embarked'].fillna('0', inplace=True)
  raw_input=raw_input.astype({'Embarked': 'int32'})
  raw_input['Cabin']=raw_input['Cabin'].apply(lambda x: 1 if not pd.isnull(x) else 0)
  raw_input=raw_input.fillna(0)
  return raw_input
```

```
def MakeInputs(raw_input, x_index):
  raw_input=CleanData(raw_input)
  input=raw_input.to_numpy()
  Y_output=input[:,0]
  X_input=input[:,x_index:]
  return X_input, Y_output;
```

```
X_train, Y_train=MakeInputs(raw_input, 1)
X_test, Y_test=MakeInputs(raw_input_test, 0)
```

```
inputs=Input(shape=(8,1))
X=Dense(24, activation='relu')(inputs)
X=Dense(12, activation='relu')(X)
X=Dense(8, activation='relu')(X)
X=Flatten()(X)
X=Dense(1, activation='sigmoid')(X)
model=Model(inputs=inputs, outputs=X)
model.summary()
```

```
model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```
model.fit(X_train, Y_train, epochs=100, batch_size=16, verbose=2)
```

```
prediction = model.predict(X_test)
```

```
result= np.array(prediction)
```

```
result=np.rint(result).astype(int)
```

```
pd.DataFrame(result).to_csv("/content/drive/My Drive/kaggle/input_data/output.csv")
```
